```python
import os
from dotenv import load_dotenv
load_dotenv()  #load all the environment variables
```

```python
# Calling the token
os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings=OpenAIEmbeddings(model="text-embedding-3-large")  # So, we will be using this model.
embeddings
```

```python
text="This is a tutorial on OPENAI embedding"
query_result=embeddings.embed_query(text)  #Apply the model on our text. It will generate the vector. 
#The length of vector dimension is based on the model we selected.
query_result
```

```python
len(query_result)
```

```python
embeddings_1024=OpenAIEmbeddings(model="text-embedding-3-large",dimensions=1024)  # I can set my own dimesion also on my model.
embeddings_1024
```

```python
text="This is a tutorial on OPENAI embedding"
query_result=embeddings_1024.embed_query(text)
len(query_result)
```

```python
query_result
```

```python
# Step 1 : Get Data - Data Ingestion
from langchain_community.document_loaders import TextLoader

loader=TextLoader('../dummy_data/speech.txt')
docs=loader.load()
docs
```

```python
# Step 2 - Data Transformation
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_documents=text_splitter.split_documents(docs)
final_documents
```

```python
# Step 3 and Step 4 - 
## Vector Embedding And Vector StoreDB
from langchain_community.vectorstores import Chroma

db=Chroma.from_documents(final_documents,embeddings_1024) # Apply my vector embeddings on all my finaldocuments
db  # This is our vector store DB
```

```python
### Retrieve the results from query vectorstore db
query="It will be all the easier for us to conduct ourselves as belligerents"
retrieved_results=db.similarity_search(query)    ## this is our vector store db. We will be doing some similarity search.
print(retrieved_results)
```
